```python
from kafka import KafkaConsumer, TopicPartition
```

```python
topic = 'movielog12'
partition = 0
tp = TopicPartition(topic, partition)
```

```python
consumer = KafkaConsumer(
    topic,
    bootstrap_servers = ['localhost:9092'],
    auto_offset_reset = 'earliest'
)
```

```python
dfRaters = pd.DataFrame()

```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
for message in consumer:
    value = message.value.decode()
    splits = value.split('/')
    
    if len(splits) == 3:
        movie = splits[2]
        userSplits = splits[0].split(',')
        user = userSplits[1]
        
        
        rating = movie[len(movie) - 1]
        movie = movie[:len(movie) - 2]
        
        dfRaters = dfRaters.append({'user_id':user, 
                                    'movie': movie,
                                    'rating': rating}, 
                                   ignore_index=True)
        #print(dfRaters.head())
        #break
        
        if len(dfRaters)%5000 == 0:
            print(len(dfRaters))
            print(value)
            #print(dfRaters.head())
            #dfRaters = pd.DataFrame.from_dict(raters, orient='index')
            dfRaters.to_csv('ratings.csv', index=True)
        
```
